Now create a "clean_price" function which takes a price as argument. 

Then if this element contains $, remove the dollar (put it into a float format).
<br>Otherwise do nothing. 

```python
def clean_price(price):
    if '$' in price :
        price= price.strip("$") 
        return float(price)
    else: 
        price = np.nan
   
```

```python
## Your code here ##
# Define a preproc function to clean the dataframe
def preproc (df):
    # apply clean_price to price and mrp columns
    df['price'] = df['price'].apply(clean_price)
    df['mrp'] = df['mrp'].apply(clean_price)
    
    # compute the median of each column
    price_med = df['price'].median()
    mrp_med = df['mrp'].median()
    review_med = df['review_count'].median()
    rating_med = df['rating'].median()
    
    # using fillna method to impute missing values
    df['review_count'].fillna(review_med, inplace = True)
    df['rating'].fillna(rating_med, inplace = True)
    df['price'].fillna(price_med, inplace = True)
    df['mrp'].fillna(mrp_med, inplace = True)
    
    return df
```

#### Question 2

What are the cheapest products which exists in size 36B? In which colors are they available?

```python
#extract the rows containg 36B with df.str.contains method.
#here can also use df['available_size']== '36B'
#but if there is a list of string like ['36B','35C','37B',...], contains method is better
df_36b = df[df['available_size'].str.contains('36B')]
#extract cheapest product in 36B
df_36b_min = df_36b[df_36b['price']==df_36b['price'].min()]
#and display the color
df_36b_min[['product_name','price','available_size','color']]
```

#### Question 6

What is the most common color of the best rated products?

What is then the best rated product available in the most popular color?

```python
# extract the products with the highest rating
df_rating_max = df[df['rating'] == df['rating'].max()]
# look for the most common color by value_counts method
# then, extract the color with the highest count number
df_commoncolor = df_rating_max['color'].value_counts().idxmax()
# using str.contains method to extract the products available in the most common color from the df_rating_max which contains the products with the highest rating
df_best_pro = df_rating_max.loc[df['color'].str.contains(df_commoncolor)]

df_best_pro
```
